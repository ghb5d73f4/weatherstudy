```python
# imports
%matplotlib inline
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2010_2019.csv") 
df.info()
```

```python
df[['Thigh', 'Tmid', 'Tlow']].plot()
```

```python
plt.figure(figsize=(16, 13))
plt.grid(True)

plt.plot(df.Thigh, df.Phigh, '-',color='grey')
plt.plot(df.Thigh, df.Phigh, '.k')
plt.show()
```

```python
TvsP = np.histogram2d(df.Thigh, df.Phigh, range=[[-10, 40],[980, 1050]],
                     bins=20)
```

```python
plt.pcolormesh(TvsP[1], TvsP[2], TvsP[0].T)
```

```python
plt.figure(figsize=(16, 13))
plt.grid(True)
plt.plot(df.Thigh, df.Phigh, '-',color='grey', alpha=0.3)
plt.plot(list( [ x + np.random.uniform(-0.5, 0.5) for x in df.Thigh]), 
         list( [y + np.random.uniform(-0.5, 0.5) for y in df.Phigh]),
         '.k')
plt.contourf(TvsP[1][:-1], TvsP[2][:-1], TvsP[0].T,
           levels=6, cmap="RdBu_r")
plt.show()
```
